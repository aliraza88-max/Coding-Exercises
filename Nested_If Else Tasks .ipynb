# Number Analysis


### User se number lo aur check karo:

# Agar number positive hai:
       Agar even hai → print "Positive Even"
       warna → "Positive Odd"
# Agar negative hai:
       Agar even hai → "Negative Even"
       warna → "Negative Odd"
       Agar 0 hai → "Zero" 

```python
num = int(input("Enter Your Number"))
if num > 0:
    if num % 2 == 0:
        print("Positive Even")
        
    else:
        print("Positive Odd")
elif  num < 0:
    if  num % 2 == 0:
        print("negative Even")
                
    else:
        print("Negative Odd")
elif num == 0:
    print("Zero")
        
    
```

# ATM Withdrawal

## User se balance aur withdraw amount lo:

    Agar amount <= balance:
    Agar amount % 500 == 0 → "Transaction Successful"
    warna → "Enter amount in multiples of 500"
    warna → "Insufficient Balance"


```python
balance=int(input("Enter Your Balance:"))
withdraw=int(input("Enter amount you wan't to withdraw:"))

if withdraw <= balance:
    if withdraw % 500 == 0:
        print("Transaction Successful")
    else:
        print("Enter amount in multiple of 500")
else:
    print("Insufficient balance")
    
```
